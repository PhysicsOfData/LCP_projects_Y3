```python
#MAGNETOMETRO
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(10.5, 3))

ax1.plot(df['Time'],df['MagnX'])
ax2.plot(df['Time'],df['MagnY'])
ax3.plot(df['Time'],df['MagnZ'])

ax1.set_title('MagnX')
ax2.set_title('MagnY')
ax3.set_title('MagnZ')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(15, 3))

FFTaccX=np.fft.fft(df['AccX'].iloc[1000:11000]-np.mean(df['AccX'].iloc[1000:11000]))
freq = np.fft.fftfreq(d=1/200, n=len(FFTaccX))
ax1.plot(freq,np.abs(FFTaccX))
#ax1.set_xlim(-0.3,20)

FFTaccY=np.fft.fft(df['AccY'].iloc[1000:11000]-np.mean(df['AccY'].iloc[1000:11000]))
ax2.plot(freq,np.abs(FFTaccY))
#ax2.set_xlim(-0.3,20)

FFTaccZ=np.fft.fft(df['AccZ'].iloc[1000:11000]-np.mean(df['AccZ'].iloc[1000:11000]))
ax3.plot(freq,np.abs(FFTaccZ))
#ax3.set_xlim(-0.3,20)

ax1.set_title('spettro AccX')
ax2.set_title('spettro AccY')
ax3.set_title('spettro AccZ')
```
